```python
from main import simulate, processing_only

# Tube
system_cfg_path = 'cfg_files/tests/tube.cfg'
simulation_cfg_path = 'cfg_files/tests/simulation_tube.cfg'
processing_simulation_path = 'cfg_files/tests/processing_tube.cfg'
```
